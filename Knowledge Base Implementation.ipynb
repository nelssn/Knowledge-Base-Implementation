This notebook demonstrates a simple knowledge representation system using structured facts, logical rules, and inference to derive new knowledge and answer queries.


```python
# Knowledge Representation System

# Facts stored as (subject, relation, object) triples
facts = [
    ("Rain", "causes", "Wet ground"),
    ("Wet ground", "causes", "Slippery roads"),
    ("Slippery roads", "increase_risk_of", "Accidents")
]

# Rules for inference
rules = [
    {
        "if": ("Rain", "causes", "Wet ground"),
        "then": ("Wet ground", "results_in", "Traffic delays")
    },
    {
        "if": ("Wet ground", "causes", "Slippery roads"),
        "then": ("Slippery roads", "leads_to", "Slow traffic")
    }
]
```

```python
# Apply rule-based inference
inferred_facts = []

for rule in rules:
    if rule["if"] in facts:
        inferred_facts.append(rule["then"])

# Combine original and inferred facts
all_facts = facts + inferred_facts

print("All Inferred Facts:")
for fact in inferred_facts:
    print(fact)
```

```python
# Query function to retrieve facts
def query(subject=None, relation=None, obj=None):
    results = []
    for fact in all_facts:
        s, r, o = fact
        if (subject is None or subject == s) and \
           (relation is None or relation == r) and \
           (obj is None or obj == o):
            results.append(fact)
    return results

# Example query
query_result = query(subject="Wet ground")

print("\nQuery Results:")
for result in query_result:
    print(result)
```
